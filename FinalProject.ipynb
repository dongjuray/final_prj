# 데이터 로드

```python
df=pd.read_csv('/content/drive/MyDrive/LIKELION_데이터분석/finalproject/globalterrorism_2013_after.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# 1. 결측치 비율 계산
missing_ratio = df.isnull().mean()

# 2. 비율이 0.8 이상인 컬럼 리스트
cols_to_drop = missing_ratio[missing_ratio >= 0.8].index

# 3. 해당 컬럼 제거
df = df.drop(columns=cols_to_drop)
```

```python
print(f"삭제된 컬럼 수: {len(cols_to_drop)}개")
print("삭제된 컬럼 이름:", list(cols_to_drop))
print(df.shape)
```

```python
df.info()
```

```python
columns_to_drop = ['country', 'region', 'summary', 'targtype1', 'targsubtype1', 'natlty1']
df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
df.shape
```

```python
columns_to_drop = ['weaptype1','weapsubtype1','propextent']
df=df.drop(columns=columns_to_drop)
df.shape
```

```python
columns_to_drop = ['scite1','scite2','scite3']
df=df.drop(columns=columns_to_drop)
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
columns_to_drop = ['location']
df=df.drop(columns=columns_to_drop)
df.shape
```

```python
df = df.dropna(subset=['latitude', 'longitude'])
```

```python
df.shape
```

```python
columns_to_drop = ['addnotes']
df=df.drop(columns=columns_to_drop)
df.isnull().sum()
```

```python
columns_to_drop = ['weapdetail']
df=df.drop(columns=columns_to_drop)
df.isnull().sum()
```

```python
df_num = df.select_dtypes(include='number')
df_num.isnull().sum()
```

```python
df.to_csv('terror_cleaned.csv',index=False, encoding='utf-8-sig')
```
